```
import tensorflow as tf
from tensorflow import keras
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score
from keras import metrics
```

```
(train_X, train_y),(test_X, test_y) = keras.datasets.boston_housing.load_data(test_split=0.2)
```

```
print(train_X.shape)
print(train_y.shape)
print(test_X.shape)
print(test_y.shape)
```

```
activation = tf.keras.activations.relu
optim = tf.keras.optimizers.Adam
metric = tf.keras.metrics.mse

model = keras.Sequential([
                          #keras.layers.Flatten(input_shape=(13,)),
                          keras.layers.Dense(10,input_dim=train_X.shape[1],activation=activation),
                          keras.layers.Dense(128,activation=activation),
                          keras.layers.Dense(64,activation=activation),
                          keras.layers.Dense(32,activation=activation),
                          keras.layers.Dense(10),
                          keras.layers.Dense(1)
])
```

```
model.compile(optimizer='rmsprop', metrics=metric,loss=tf.keras.losses.mean_squared_error)
```

```
model.summary()
```

```
model.fit(x=train_X,y=train_y,epochs=2000,validation_data=(test_X,test_y), batch_size=128)
```

```
test_mse_score, test_mae_score = model.evaluate(test_X, test_y)

print(test_mse_score)
print(test_mae_score)
```
